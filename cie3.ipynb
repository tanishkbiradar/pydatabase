```python
import pandas as pd
```

```python
df1 = pd.read_csv('dataset/bank-additional-full.csv')
df1
```

```python
df1.drop(['duration'],axis=1,inplace=True)
```

```python
df1.sample().shape
```

```python
X = df1.drop(['y'],axis=1)
X
```

```python
Y = df1['y']
Y
```

```python
get_dm = pd.get_dummies(X,drop_first=True)
get_dm.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(get_dm , Y , train_size=0.8 , random_state=2)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
```

```python
Y_pred = dtc.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score
```

```python
accuracy_score(Y_test , Y_pred)
```

```python
precision_score(Y_test , Y_pred , average='macro')
```

```python
recall_score(Y_test , Y_pred , average='micro')
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc1 = DecisionTreeClassifier(splitter='random',criterion='entropy')
dtc1.fit(X_train,Y_train)
```

```python
Y_pred1 = dtc1.predict(X_test)
Y_pred1
```

```python
accuracy_score(Y_test , Y_pred1)
```

```python
precision_score(Y_test , Y_pred1 , average='macro')
```

```python
recall_score(Y_test , Y_pred1 , average='micro')
```
