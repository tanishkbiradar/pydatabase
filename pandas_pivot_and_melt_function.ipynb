```python
import pandas as pd
from pandas import DataFrame
df1 = DataFrame([
                 ["Aryan","Singh","Full Time","administration",2,20000],
                 ["Rohan","Agarwal","Intern","Technical",3,5000],
                 ["Riya","Shah","Full Time","administration",5,10000],
                 ["Yash","Bhatia","Part Time","Technical",7,10000],
                 ["Siddant","Khanna","Full Time","Management",6,20000]
    
                 
               ],columns = ["First Name","Last Name","Type","Department","yoE","Salary"])
df1
```

```python
a = pd.pivot_table(df1,index="Type",columns="Department")
a
```

```python
a = pd.pivot_table(df1,index="Type",columns="Department",values="Salary")
a
```

```python
a = pd.pivot_table(df1,index="Type")
a
```

```python
a = pd.pivot_table(df1,index="Type",columns="Department",aggfunc="sum")
a
```
