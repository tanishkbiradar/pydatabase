```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('dataset/Iris.csv')
```

```python
df
```

```python
x = df.iloc[:,0:4]
x
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
si = StandardScaler()
```

```python
x = pd.DataFrame(si.fit_transform(x),columns = x.columns)
x
```

```python
y = df.iloc[:,-1]
y
```

```python
x.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(x.corr())
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 4)
x_pca = pd.DataFrame(pca.fit_transform(x),columns = ['p1','p2','p3','p4'])
x_pca
```

```python
sns.heatmap(x_pca.corr())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_pca,y,test_size = 0.8,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
