```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset/demo.csv")
df
```

```python
df1 = df.iloc[:,1:2]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['DIST','FC','SC','pass','Fail']])
```

```python
oe.fit(df1)
```

```python
df2 = oe.transform(df1)
```

```python
df2
```

```python
df3 = pd.DataFrame(df2,columns=df1.columns)
df3
```
