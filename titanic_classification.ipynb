```python
df.fillna(method = 'bfill',inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns = ['PassengerId','Name','Ticket','Fare','Cabin'],inplace = True)
df
```

```python
x = pd.DataFrame(df.iloc[:,1:])
```

```python
x
```

```python
y = df.iloc[:,0]
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
tr = ColumnTransformer(transformers=[
    ('t1',OneHotEncoder(drop = 'first',sparse = False),['Sex','Embarked'])
],remainder='passthrough')
```

```python
x1=tr.fit_transform(x)
x1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y,test_size = 0.2,random_state = 2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
a = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print(accuracy_score(y_test,a))
```

```python
pd.DataFrame(confusion_matrix(y_test,a))
```

```python
result = pd.DataFrame()
result['Actual label'] = y_test
result['Decision Tree Prediction'] = a
result
```
